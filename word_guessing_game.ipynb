```python
import random

# Function to select a random word from words.txt
def get_word():
    with open("words.txt", "r") as f:
        words = f.readlines()
    return random.choice(words).strip()

# Function to display the word with underscores for unguessed letters
def display_word(word, guesses):
    display = ""
    for letter in word:
        if letter in guesses:
            display += letter + " "
        else:
            display += "_ "
    return display

# Function to get a guess from the player
def get_guess():
    print("\n")
    guess = input("Guess a letter: ")
    if len(guess) != 1:
        print("Please enter a single letter.")
        return get_guess()
    elif not guess.isalpha():
        print("Please enter a letter.")
        return get_guess()
    else:
        return guess

# Main function to run the game
def play_game():
    word = get_word()
    guesses = set()
    incorrect_guesses = 0
    max_incorrect_guesses = 8

    while True:
        print(display_word(word, guesses))
        if "_" not in display_word(word, guesses):
            print("\n")
            print("Congratulations, you guessed the word!")
            break
        elif incorrect_guesses == max_incorrect_guesses:
            print("\n")
            print(f"Sorry, you ran out of guesses. The word was {word}.")
            break
        else:
            guess = get_guess()
            if guess in guesses:
                print("\n")
                print("You already guessed that letter. Try again.")
            elif guess in word:
                guesses.add(guess)
                print("\n")
                print("Good guess!")
            else:
                guesses.add(guess)
                incorrect_guesses += 1
                print("\n")
                print(f"Sorry, {guess} is not in the word. You have {max_incorrect_guesses - incorrect_guesses} incorrect guesses left.")

if __name__ == '__main__':
    play_game()
```
